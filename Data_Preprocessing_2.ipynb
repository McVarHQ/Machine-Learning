```python
raw_data = {'first_name': ['Jason', np.nan, 'Tina', 'Jake', 'Amy'],
        'last_name': ['Miller', np.nan, 'Ali', 'Milner', 'Cooze'],
        'age': [42, np.nan, 36, 24, 73],
        'sex': ['m', np.nan, 'f', 'm', 'f'],
        'preTestScore': [4, np.nan, np.nan, 2, 3],
        'postTestScore': [25, np.nan, np.nan, 62, 70]}
df = pd.DataFrame(raw_data, columns = ['first_name', 'last_name', 'age', 'sex', 'preTestScore', 'postTestScore'])

df
```

```python
df.dropna(how='all',axis=0)  #removes
df

```

```python
df['location']=np.nan
df
```

```python
url='https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv'
data=pd.read_csv(url)
```

```python
data.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data
```

```python
#printing boxplot
sns.boxplot(data.dis)
plt.show()
```

```python
l=list(data.columns.values)
```

```python
#print IQR for each column
for i in l:
  val=(np.percentile(data[i],75)-np.percentile(data[i],25)).round(2)
  print(i,val)
```

```python
t=data[['crim','zn','indus','chas','nox','rm','age','dis','tax','ptratio','b','lstat','medv']].corr()
sns.heatmap(t,annot=True,cmap='Greens')
```
